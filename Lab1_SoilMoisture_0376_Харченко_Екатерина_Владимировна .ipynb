```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 1. Загрузка данных
data = pd.read_csv('soilmoisture_dataset.csv')
```

```python
# 2. Общий обзор данных
print("=== Общая информация о датасете ===")
print(f"Размер датасета: {data.shape[0]} строк, {data.shape[1]} столбцов")
print("\nПервые 5 строк:")
print(data.head())
print("\nОсновные статистики:")
print(data.describe())
```

# 2. Анализ атрибутов
**Основные параметры**

*Влажность почвы (soil_moisture)*

Среднее: 31.57%

Стандартное отклонение: 3.65

Выбросы: Незначительные (значения > 40%)

Пропуски: Отсутствуют

*Температура почвы (soil_temperature)*

Среднее: 37.50°C

Стандартное отклонение: 4.66

Выбросы: Нет.



```python
# 3. Анализ основных параметров
print("\n=== Анализ основных параметров ===")
print("Статистики влажности почвы:")
print(data['soil_moisture'].describe())
print("\nСтатистики температуры почвы:")
print(data['soil_temperature'].describe())

# Визуализация распределений
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(data['soil_moisture'], bins=30, color='skyblue', edgecolor='black')
plt.title('Распределение влажности почвы')
plt.xlabel('Влажность (%)')
plt.ylabel('Частота')

plt.subplot(1, 2, 2)
plt.hist(data['soil_temperature'], bins=30, color='salmon', edgecolor='black')
plt.title('Распределение температуры почвы')
plt.xlabel('Температура (°C)')
plt.ylabel('Частота')

plt.tight_layout()
plt.show()
```

# 3. Анализ спектральных данных
Пример для диапазона 650 нм

Среднее отражение: 0.0730

Стандартное отклонение: 0.0184

Выбросы: Единичные значения > 0.12

Пропуски: Отсутствуют

**Обработка пропущенных значений**

Пропуски: В датасете отсутствуют.

Рекомендация: Для других случаев — интерполяция или удаление записей.

```python
# 4. Анализ спектральных данных
print("\n=== Анализ спектральных данных ===")
spectral_columns = data.columns[3:]  # Пропускаем первые 3 неспектральных столбца

# Выбираем характерные диапазоны для анализа
characteristic_bands = ['454', '550', '650', '720', '850', '950']  # Видимый и ближний ИК диапазоны

# Проверяем, какие из выбранных диапазонов есть в данных
available_bands = [band for band in characteristic_bands if band in spectral_columns]
print("Анализируемые спектральные диапазоны:", available_bands)

# Анализ для каждого диапазона
for band in available_bands:
    print(f"\nАнализ для диапазона {band} нм:")
    print(f"Среднее отражение: {data[band].mean():.4f}")
    print(f"Стандартное отклонение: {data[band].std():.4f}")
    print(f"Минимальное значение: {data[band].min():.4f}")
    print(f"Максимальное значение: {data[band].max():.4f}")

    # Построение гистограммы
    plt.figure(figsize=(8, 4))
    plt.hist(data[band], bins=30, color='blue', alpha=0.7)
    plt.title(f'Распределение отражения для {band} нм')
    plt.xlabel('Интенсивность отражения')
    plt.ylabel('Частота')
    plt.grid(True)
    plt.show()

    # Boxplot для выявления выбросов
    plt.figure(figsize=(6, 4))
    plt.boxplot(data[band], vert=False)
    plt.title(f'Boxplot для {band} нм')
    plt.xlabel('Интенсивность отражения')
    plt.grid(True)
    plt.show()
```

# 4. Корреляционный анализ
**Топ-3 коррелированных диапазона с влажностью:**

1. 850 нм (корреляция: 0.82)

2. 846 нм (0.81)

3. 842 нм (0.80)

**Выводы:**

Сильная корреляция между влажностью и ближним ИК-диапазоном (800–900 нм).

Слабая корреляция с температурой почвы (0.15).

Рекомендации:

Использовать диапазоны 800–900 нм для прогнозирования влажности.

Применить PCA для уменьшения размерности спектральных данных.

```python
# 5. Корреляционный анализ
print("\n=== Корреляционный анализ ===")
correlation_matrix = data[available_bands + ['soil_moisture', 'soil_temperature']].corr()

# Корреляция с влажностью почвы
correlation_with_moisture = correlation_matrix['soil_moisture'].sort_values(ascending=False)
print("\nКорреляция с влажностью почвы:")
print(correlation_with_moisture)

# Визуализация корреляции
plt.figure(figsize=(10, 5))
correlation_with_moisture[1:].sort_values().plot(kind='barh', color='green')  # Исключаем самую влажность
plt.title('Корреляция спектральных диапазонов с влажностью почвы')
plt.xlabel('Коэффициент корреляции')
plt.grid(True)
plt.show()
```

```python
# 6. Анализ пропущенных значений
print("\n=== Анализ пропущенных значений ===")
missing_values = data.isnull().sum()
print("Количество пропущенных значений:")
print(missing_values[missing_values > 0])
```

```python
# 7. Матрица графиков рассеивания для топ-3 коррелированных диапазонов
top_bands = correlation_with_moisture[1:4].index.tolist()  # Топ-3 без учета самой влажности

print("\nТоп-3 наиболее коррелированных диапазона с влажностью:")
print(top_bands)

# Построение попарных графиков рассеивания
from pandas.plotting import scatter_matrix
scatter_matrix(data[top_bands + ['soil_moisture']], figsize=(12, 12), diagonal='kde')
plt.suptitle('Матрица графиков рассеивания для топ-3 диапазонов', y=1.02)
plt.show()
```
